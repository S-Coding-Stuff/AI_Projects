```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jrobischon/wikipedia-movie-plots")
```

```python
import pandas as pd, numpy as np

movies = pd.read_csv(f'{path}/wiki_movie_plots_deduped.csv')
```

```python
movies['Origin/Ethnicity'].value_counts().reset_index().sort_values('count', ascending=False)
```

```python
movies = movies[movies['Origin/Ethnicity'].isin(['American', 'British', 'Australian', 'Canadian'])]
```

```python
movies
```

```python
movies['Title-And-Plot'] = movies['Title'] + ' : ' + movies['Plot']
```

```python
movies
```

```python
# Will need to reclassify all unknown genres
movies['Genre'].value_counts().reset_index().sort_values('count', ascending=False)
```

```python
genre_mapping = {
    'sci-fi' : 'science fiction',
    'scifi': 'science fiction',
    'crime drama' : 'crime',
    'war drama' : 'war',
    'biopic' : 'biography',
    'historical drama' : 'historical',
    'anime':'animation',
    'bio':'biography',
    'world war ii':'war',
    'documentary short':'documentary',
    'drama':'drama',
    'action':'action',
    'adventure':'adventure',
    'comedy':'comedy',
    'horror':'horror',
    'romance':'romance',
    'science fiction':'science fiction',
    'animation':'animation',
    'crime':'crime',
    'fantasy':'fantasy',
    'mystery':'mystery',
    'family':'family',
    'documentary':'documentary',
    'biography':'biography',
    'history':'history',
    'war':'war',
    'western':'western',
    'sport':'sport',
    'sports':'sport',
    'musical':'musical',
    'superhero':'superhero',
    'thriller':'thriller'
}
```

```python
movies['New-Genre'] = movies['Genre'].map(genre_mapping)
```

```python
movies
```

```python
movies['Genre-And-Plot'] = np.where(
    movies['Genre'] != 'unknown',
    movies['Genre'] + ' : ' + movies['Plot'],
    movies['Plot']
)
```

```python
movies.tail(5)
```

```python
movies['New-Genre'].value_counts().reset_index().sort_values('count', ascending=False)
```

```python
from transformers import pipeline
key_genres = [
    'drama', 'action', 'adventure', 'comedy','horror','romance','science fiction', 'animation', 'crime', 'fantasy', 'mystery','family','documentary','biography','history','war','western','sport','musical','superhero','thriller', 'other'
]

pipe = pipeline('zero-shot-classification', 'FacebookAI/roberta-large-mnli')
```

```python
#sequence = books.loc[books['categories'] == 'Fiction', 'description'].reset_index(drop=True)[0]
sequence = movies.loc[movies['New-Genre'] == 'science fiction', 'Genre-And-Plot'].reset_index(drop=True)[0]
```

```python
pipe(sequence, key_genres)
```

Using 'Genre-And-Plot' for validation as this provide better distinction of a given key genre. For example 'short documentary' will go to 'documentary'

```python
seq_2 = movies.loc[movies['New-Genre'].isna(), 'Genre-And-Plot'].reset_index(drop=True)[300]
```

```python
pipe(seq_2, key_genres)
```
